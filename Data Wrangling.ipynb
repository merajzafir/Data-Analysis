## Also called as Data Pre-processing or Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

df= pd.read_csv(filename)
df.head(5)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head(2)
```

## Dealing with Missing Values

```python
# Replacing '?' with NaN
df.replace('?', np.nan, inplace=True)
df.head(5)
```

```python
# Looking For Missing Values in ecah Feature/Column
df.isnull().sum()
```

```python
# Returns Not Null Values
df.notnull().sum()
```

```python
# Replace NaN by Mean value in 'normalized-losses' column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# returns unique values with their frequency
df['num-of-doors'].value_counts()
```

```python
# returns most frequently appeared values 
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

## Correcting Datatypes

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

## Data Standardization
Standardization is the process of transforming data into a common format which allows the researcher to make the meaningful comparison.

```python
# transform mpg to L/100km in the column of "highway-mpg", 
# and change the name of column to "highway-L/100km".
# Write your code below and press Shift+Enter to execute 

df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={"highway-mpg":"highway-L/100km"})
```

## Normalization
Normalization is the process of transforming values of several variables into a similar range.
Typical normalizations include scaling the variable so the variable values range from 0 to 1

```python
# Simple Feature Scaling

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height']= df['height']/df['height'].max()
# show the scaled columns
df[["length","width","height"]].head()
```

```python
# Min-Max

df['length']= (df['length']-df['length'].min()) / (df['length'].max()-df['length'].min())
```

```python
# Z-score
df['length'] = (df['length']-df['length'].mean()) / df['length'].std()
```

## Binning
Binning is the process of transforming Numeric Variable into Categorical Variable

We would like 3 bins of equal size bandwidth so we use numpy's function  
**linspace(start_value, end_value, numbers_generated) .**

Since we want to include the minimum value of horsepower we want to set start_value=min(df["horsepower"]).

Since we want to include the maximum value of horsepower we want to set end_value=max(df["horsepower"]).

Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated=4.

We build a bin array, with a minimum value to a maximum value, with bandwidth calculated above. The bins will be values used to determine when one bin ends and another begins.

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

## One-Hot Encoding
It refers to splitting the column which contains categorical data to many columns depending on the number of categories present in that column. Each column contains “0” or “1” corresponding to which column it has been placed.

```python
dummy_variable = pd.get_dummies(df['fuel-type'])
dummy_variable.head()
```

```python
# change column names for clarity
dummy_variable.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable.head()
```

```python
# Merging dataframe "df" and "dummy_variable"
df= pd.concat([df, dummy_variable], axis=1)
```

```python
# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```
